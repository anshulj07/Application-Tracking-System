```python
import re
def cleanResumeTxt(txt):
    cleanTxt = re.sub('http\S+\s', '', txt)
    cleanTxt = re.sub('@\S+', '', cleanTxt)
    cleanTxt = re.sub('#\S+\s', '', cleanTxt)
    return cleanTxt
```
